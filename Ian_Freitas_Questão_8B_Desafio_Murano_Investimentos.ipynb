```python
def Q8_B(stocks):
  import matplotlib.pyplot as plt
  import pandas as pd
  import seaborn as sns

  stocks_df = pd.DataFrame()

  for stock in stocks:
    stock_ = stock + '.us.txt'
    stocks_df[stock.upper()] = pd.read_csv(stock_)['Close']

  corr = stocks_df.corr()

  plt.subplots(figsize=(10,8))  
  sns.heatmap(corr, cmap='vlag', annot=True, vmin=-1, vmax=1)

stocks_list = ['aapl','elec','fss','ghy','googl','mgu','mjco','msft','ndsn','shak']

Q8_B(stocks_list)
```
